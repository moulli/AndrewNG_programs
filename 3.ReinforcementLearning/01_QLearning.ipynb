```python
# Taxi v2:

env = gym.make("Taxi-v2")
env.render()
```

```python
# Hyperparameters:

max_parties = 50000
test_parties = 100
max_steps = 99

alpha_rate = 0.1
gamma_rate = 0.9

epsilon = 1.0
epsilon_max = 1.0
epsilon_min = 0.01
decay_rate = 0.1
```

```python
# Defining Q table at 0

action_size = env.action_space.n
state_size = env.observation_space.n

Q_table = np.zeros((state_size, action_size))
print(Q_table.shape)
```

```python
# Training algorithm

for i in range(max_parties):
    
    # Initialization
    state = env.reset()
    
    # Algorithm for each game
    for ite in range(max_steps):

        # Exploration or exploitation
        explo = random.uniform(0, 1)
        if explo > epsilon:
            action = np.argmax(Q_table[state, :])
        else:
            action = env.action_space.sample()

        # Updating Q table
        new_state, reward, done, info = env.step(action)
        Q_table[state, action] = Q_table[state, action] + alpha_rate * (reward + gamma_rate*np.max(Q_table[new_state, :]) - Q_table[state, action])
        # Updating state
        state = new_state

        # Breaking if necessary
        if done == True:
            break

    # Decreasing epsilon
    epsilon = epsilon_min + (epsilon_max - epsilon_min) * np.exp(-decay_rate * i)
    
    # Print improvement
    if np.mod(i+1, np.floor(max_parties/10)) == 0:
        print("Iteration %.0f out of %.0f." %(i+1, max_parties))
```

```python
# Testing algorithm:

rewards = np.zeros((test_parties, 1))

for i in range(test_parties):
    
    # Initialization
    state = env.reset()
    reward_temp = 0
    
    # Main algorithm
    for ite in range(max_steps):
        action = np.argmax(Q_table[state, :])
        state, reward, done, info = env.step(action)
        reward_temp = reward_temp + reward
        if done == True:
            break
    
    # Uddating rewards
    rewards[i] = reward_temp
    
print("Mean reward is %.2f, maximum reward is %.2f, minimum reward is %.2f." %(np.mean(rewards), np.max(rewards), np.min(rewards)))
```

```python
# Plotting algorithm

state = env.reset()
for ite in range(max_steps):
    env.render()
    action = np.argmax(Q_table[state, :])
    state, reward, done, info = env.step(action)
    reward_temp = reward_temp + reward
    if done == True:
        break
env.close()
```
