```python
# Parameters
dim_plan_mm = np.array([0.496, 1.122, 0.276])
dim_plan_pix = np.array([653, 1009])
meanx = 4.5
stdx = 1
meany = 4.5
stdy = 1
covcor = 0
stdcov = 0.5
meanoise = 0.001
stdnoise = 0.002
# Parameters for the different subplans
planpar = np.array([[250.0, 100.0, 50.0, 15.0], # value for neurons
                    [50.0, 30.0, 20.0, 4.0], # variance for value for neurons
                    [0.0, 0.1, 1.5, 2.5]]) # repartitions of values for neurons
splan = planpar.shape[1]
numneur = 5000
numstd = 500
numfin = int(np.round_(numneur + numstd*np.random.randn())) # number of neurons
coordfin = np.zeros((numfin, 2)) # centers of neurons
plan = np.zeros((dim_plan_pix[0], dim_plan_pix[1], splan))
```

```python
# For-loop
Xfield, Yfield = np.mgrid[-6:7, -6:7]
Xfield = Xfield.flatten()[:, np.newaxis]
Yfield = Yfield.flatten()[:, np.newaxis]
for i in range(numfin):
    # Taking a plan at random based on probabilities from planpar
    planrand = splan * np.random.rand()
    planchoice = np.where(planrand <= np.append(planpar[2, :], splan))
    planchoice = planchoice[0][0] - 1
    # Finding a location for new neuron or going on
    finding = -1
    numround = 0
    while finding == -1 and numround < 100:
        randx = int(np.round_(np.random.rand() * (dim_plan_pix[0] - 15)) + 7)
        randy = int(np.round_(np.random.rand() * (dim_plan_pix[1] - 15)) + 7)
        anafield = plan[(randx-7):(randx+6), (randy-7):(randy+6), planchoice]
        if np.count_nonzero(anafield) == 0:
            break
        numround += 1
    if numround == 100:
        continue
    # Defining neuron's information
    matcorval = covcor + stdcov * np.random.randn()
    matrixstd = np.array([[meanx + stdx*np.random.randn(), matcorval], 
                         [matcorval, meany + stdy*np.random.randn()]])
    if np.linalg.det(matrixstd) < 0:
        continue
    else:
        coordfin[i, 0] = randx
        coordfin[i, 1] = randy
    numval = (planpar[0, planchoice] + planpar[1, planchoice]*np.random.randn()) * np.sqrt((2*np.pi)**2 * np.linalg.det(matrixstd))
    mvn = multivariate_normal(mean = meaneu, cov = matrixstd)
    valexp = numval * mvn.pdf(np.hstack((Xfield, Yfield)))
    # Replacing in final plan
    plan[(randx-7):(randx+6), (randy-7):(randy+6), planchoice] = valexp.reshape(13, 13)
```

Now we would like to create a dataset from these stacked plans. We are going to implement a few functions. The first one will return plans filled with neurons, based on some parameters. The second one is going to build a usable example based on a number of plans by stacking them and adding noise. The last one is the function that will provide a whole dataset based on initial parameters.

```python
# Function to create a plan
def create_plan(number_of_plans, plan_size, neuronnumber_mean, neuronnumber_std, neuron_mean, neuron_std, cormat_mean, cormat_std, xvar_mean, xvar_std, yvar_mean, yvar_std):
    # Create initial set of plans
    plan = np.zeros((plan_size[0], plan_size[1], number_of_plans))
    # Define number of neurons per plan
    numfin = np.round_(neuronnumber_mean + neuronnumber_std*np.random.randn(number_of_plans))
    # Define coordinates vector
    coord_neu = np.zeros((0, 3))
    # Loop on each plan
    for i in range(number_of_plans):
        for j in range(int(numfin[i])):
            finding = -1
            numround = 0
            numlim = 50
            while finding == -1 and numround < numlim:
                randx = int(np.round_(np.random.rand() * (plan_size[0] - 15)) + 7)
                randy = int(np.round_(np.random.rand() * (plan_size[1] - 15)) + 7)
                anafield = plan[(randx-7):(randx+6), (randy-7):(randy+6), i]
                if np.count_nonzero(anafield) == 0:
                    break
                numround += 1
            if numround == numlim:
                continue
            # Defining neuron's information
            matcorval = cormat_mean + cormat_std * np.random.randn()
            matrixstd = np.array([[xvar_mean + xvar_std*np.random.randn(), matcorval], 
                                 [matcorval, yvar_mean + yvar_std*np.random.randn()]])
            if np.linalg.det(matrixstd) < 0:
                continue
            else:
                coord_neu = np.vstack((coord_neu, np.array([randx, randy, i])))
            numval = (neuron_mean + neuron_std*np.random.randn()) * np.sqrt((2*np.pi)**2 * np.linalg.det(matrixstd))
            mvn = multivariate_normal(mean = meaneu, cov = matrixstd)
            valexp = numval * mvn.pdf(np.hstack((Xfield, Yfield)))
            # Replacing in final plan
            plan[(randx-7):(randx+6), (randy-7):(randy+6), i] = valexp.reshape(13, 13)          
    return plan, coord_neu
                
```

```python
def add_neurons_pool(plan_size, numneur, info_neuron):
    # Retriving information on neurons
    neuron_mean, neuron_std, cormat_mean, cormat_std, xvar_mean, xvar_std, yvar_mean, yvar_std = info_neuron
    # Creating plan and coord_neu
    plan_add = np.zeros((plan_size[0], plan_size[1]))
    coord_neu_add = np.zeros((0, 2))
    # Max number of iterations to place new neuron
    numlim = 50
    # Main for-loop
    for j in range(int(numneur)):
        finding = -1
        numround = 0
        while finding == -1 and numround < numlim:
            randx = int(np.round_(np.random.rand() * (plan_size[0] - 15)) + 7)
            randy = int(np.round_(np.random.rand() * (plan_size[1] - 15)) + 7)
            anafield = plan_add[(randx-7):(randx+6), (randy-7):(randy+6)]
            if np.count_nonzero(anafield) == 0:
                break
            numround += 1
        if numround == numlim:
            continue
        # Defining neuron's information
        matcorval = cormat_mean + cormat_std * np.random.randn()
        matrixstd = np.array([[xvar_mean + xvar_std*np.random.randn(), matcorval], 
                             [matcorval, yvar_mean + yvar_std*np.random.randn()]])
        if np.linalg.det(matrixstd) < 0:
            continue
        else:
            coord_neu_add = np.vstack((coord_neu_add, np.array([randx, randy])))
        numval = (neuron_mean + neuron_std*np.random.randn()) * np.sqrt((2*np.pi)**2 * np.linalg.det(matrixstd))
        mvn = multivariate_normal(mean = meaneu, cov = matrixstd)
        valexp = numval * mvn.pdf(np.hstack((Xfield, Yfield)))
        # Replacing in final plan
        plan_add[(randx-7):(randx+6), (randy-7):(randy+6)] = valexp.reshape(13, 13)  
    return plan_add, coord_neu_add

def plan_pooling(number_of_plans, plan_size, neuronnumber_mean, neuronnumber_std, info_neuron):
    # Retriving information on neurons
    neuron_mean, neuron_std, cormat_mean, cormat_std, xvar_mean, xvar_std, yvar_mean, yvar_std = info_neuron
    # Create initial set of plans
    plan = np.zeros((plan_size[0], plan_size[1], number_of_plans))
    # Define number of neurons per plan
    numfin = np.round_(neuronnumber_mean + neuronnumber_std*np.random.randn(number_of_plans))
    # Define coordinates vector
    coord_neu = np.zeros((0, 3))
    # Define pools
    pool = mp.Pool(processes = number_of_plans)
    results = [pool.apply(add_neurons_pool, args = (plan_size, numfin[i], info_neuron)) for i in range(number_of_plans)]
    for i in range(number_of_plans):
        plan[:, :, i] = results[i][0]
        coord_neu = np.vstack((coord_neu, np.hstack((results[i][1], i*np.ones((results[i][1].shape[0], 1))))))
    return plan, coord_neu
```

```python
def create_plans(number_of_plans, plan_size, neuronnumber_mean, neuronnumber_std, info_neuron):
    # Retriving information on neurons
    neuron_mean, neuron_std, cormat_mean, cormat_std, xvar_mean, xvar_std, yvar_mean, yvar_std = info_neuron
    # Create initial set of plans
    plan = np.zeros((plan_size[0], plan_size[1], number_of_plans))
    # Define number of neurons per plan
    numfin = np.round_(neuronnumber_mean + neuronnumber_std*np.random.randn(number_of_plans))
    # Define coordinates vector
    coord_neu = np.zeros((0, 3))
    # Loop on each plan
    for i in range(number_of_plans):
        for j in range(int(numfin[i])):
            finding = -1
            numround = 0
            numlim = 50
            while finding == -1 and numround < numlim:
                randx = int(np.round_(np.random.rand() * (plan_size[0] - 15)) + 7)
                randy = int(np.round_(np.random.rand() * (plan_size[1] - 15)) + 7)
                anafield = plan[(randx-7):(randx+6), (randy-7):(randy+6), i]
                if np.count_nonzero(anafield) == 0:
                    break
                numround += 1
            if numround == numlim:
                continue
            # Defining neuron's information
            matcorval = cormat_mean + cormat_std * np.random.randn()
            matrixstd = np.array([[xvar_mean + xvar_std*np.random.randn(), matcorval], 
                                 [matcorval, yvar_mean + yvar_std*np.random.randn()]])
            if np.linalg.det(matrixstd) < 0:
                continue
            else:
                coord_neu = np.vstack((coord_neu, np.array([i, randx-1, randy-1])))
            numval = (neuron_mean + neuron_std*np.random.randn()) * np.sqrt((2*np.pi)**2 * np.linalg.det(matrixstd))
            mvn = multivariate_normal(mean = meaneu, cov = matrixstd)
            valexp = numval * mvn.pdf(np.hstack((Xfield, Yfield)))
            # Replacing in final plan
            plan[(randx-7):(randx+6), (randy-7):(randy+6), i] = valexp.reshape(13, 13)          
    return plan, coord_neu
```

Now we are going to make a function that will create datasets by cropping, rotating and stacking plans randomly from former created distribution. This will allow us to have a complete dataset.

```python
def stack_plans(nstacks, plans_dataset, coord_dataset, stack_mean, stack_std, cropped_size, noise_mean, noise_std):
    # First defining final matrices
    stack_plan = np.zeros((nstacks, cropped_size, cropped_size))
    stack_coord = np.zeros((0, 3))
    # Number of stacks in plans_dataset
    nstacks_old = plans_dataset.shape[2]
    # Parameters for cropping
    xcrop = plans_dataset.shape[0] - cropped_size
    ycrop = plans_dataset.shape[1] - cropped_size
    # Filter for noise
    filter_shape = 5
    filter_noise = np.zeros((filter_shape, filter_shape))
    for i in range(filter_shape):
        for j in range(filter_shape):
            radius_temp = np.sqrt((i-(filter_shape-1)/2)**2 + (j-(filter_shape-1)/2)**2)
            if radius_temp <= 1:
                filter_noise[i, j] = 1
            elif radius_temp <= 2:
                filter_noise[i, j] = 0.5  
    # Launching loop
    for i in range(nstacks):
        # Selecting number of stacks
        stack_f = int(np.round_(stack_mean + stack_std*np.random.randn()))
        if stack_f <= 0:
            stack_f = 1
        # Creating temporary plan
        plans_temp = np.zeros((cropped_size, cropped_size, stack_f+1))
        # Creating a temporaty coordinates matrix
        coords_temp = np.zeros((0, 2))
        # Chosing plans to use
        plan_use = np.floor(nstacks_old*np.random.rand(stack_f)).astype(int)
        # Second for loop for each plan
        for j in range(stack_f):
            # Loading plan and coordinates
            plan_temp = plans_dataset[:, :, plan_use[j]]
            coord_temp = coord_dataset[coord_dataset[:, 0] == plan_use[j], 1:3]
            # Cropping
            xcrop_init = random.randint(0, xcrop) #np.random.random_integers(0, xcrop)
            ycrop_init = random.randint(0, ycrop) #np.random.random_integers(0, ycrop)
            coord_temp = coord_temp - np.array([[xcrop_init, ycrop_init]])
            coord_temp = coord_temp[coord_temp[:, 0] < cropped_size, :]
            coord_temp = coord_temp[coord_temp[:, 0] >= 0, :]
            coord_temp = coord_temp[coord_temp[:, 1] < cropped_size, :]
            coord_temp = coord_temp[coord_temp[:, 1] >= 0, :]
            plan_temp = plan_temp[xcrop_init:(xcrop_init+cropped_size), ycrop_init:(ycrop_init+cropped_size)]          
            # Rotating
            rotnum = random.randint(0, 3) #np.random.random_integers(0, 3)
            for k in range(rotnum):
                plan_temp = plan_temp.T[::-1, :]
                coord_temp = np.vstack((cropped_size-1-coord_temp[:, 1], coord_temp[:, 0])).T 
            # Adding to temporary stack
            plans_temp[:, :, j] = plan_temp
            coords_temp = np.vstack((coords_temp, coord_temp))
        # Adding noise plan
        plan_noise = noise_mean + np.abs(noise_std*np.random.randn(cropped_size, cropped_size))
        plans_temp[:, :, -1] = convolve(plan_noise, filter_noise, mode="constant") # convolution to filter
        # Merging stacks into stack_plan
        stack_plan_temp = np.max(plans_temp, axis=2)
        stack_plan_temp[stack_plan_temp < 0] = 0
        stack_plan[i, :, :] = stack_plan_temp
        # Deleting doublons in stack_coord
        temp_unique = np.unique(coords_temp, axis=0)
        temp_unique = np.hstack((i*np.ones((temp_unique.shape[0], 1)), temp_unique))
        stack_coord = np.vstack((stack_coord, temp_unique))
    # Returning values
    return stack_plan, stack_coord
```

Now we are going to build a training set and a test set, and train a CNN to be able to obtain centers of neurons from a simple picture

```python
# Building training and test sets
X_train = np.zeros((0, crop_size, crop_size))
Y_train = np.zeros((0, crop_size*crop_size))
X_test = np.zeros((0, crop_size, crop_size))
Y_test = np.zeros((0, crop_size*crop_size))
lensk = sk_plan.shape[0]
test_switch = int(np.round_(0.8*lensk))
for i in range(lensk):
    sk_coord_temp = sk_coord[sk_coord[:, 0] == i, 1:3]
    Y_temp = np.zeros((crop_size, crop_size))
    Y_temp[sk_coord_temp[:, 0].astype(int), sk_coord_temp[:, 1].astype(int)] = 1
    if i < test_switch:
        Y_train = np.vstack((Y_train, Y_temp.reshape(1, crop_size*crop_size)))
        X_train = np.concatenate((X_train, sk_plan[i, :, :][np.newaxis, :, :]), axis=0)
    else:
        Y_test = np.vstack((Y_test, Y_temp.reshape(1, crop_size*crop_size)))
        X_test = np.concatenate((X_test, sk_plan[i, :, :][np.newaxis, :, :]), axis=0)
# Reshaping
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], X_train.shape[2], 1)
Y_train = Y_train.reshape(Y_train.shape[0], 1, 1, Y_train.shape[1])
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], X_test.shape[2], 1)
Y_test = Y_test.reshape(Y_test.shape[0], 1, 1, Y_test.shape[1])
```

```python
# Building neural network
sk_in = Input(shape=(crop_size, crop_size, 1))
x = Conv2D(32, (3, 3), strides=(1, 1), padding="same", activation="relu", name="1st_conv")(sk_in)
x = Conv2D(32, (3, 3), strides=(1, 1), padding="same", activation="relu", name="2nd_conv")(x)
x = MaxPooling2D((2, 2), strides=(2, 2), padding="valid", name="1st_pool")(x)
x = Conv2D(64, (3, 3), strides=(1, 1), padding="same", activation="relu", name="3rd_conv")(x)
x = Conv2D(64, (3, 3), strides=(1, 1), padding="same", activation="relu", name="4th_conv")(x)
x = Conv2D(64, (3, 3), strides=(1, 1), padding="same", activation="relu", name="5th_conv")(x)
x = MaxPooling2D((2, 2), strides=(2, 2), padding="valid", name="2nd_pool")(x)
x = Conv2D(128, (3, 3), strides=(1, 1), padding="same", activation="relu", name="6th_conv")(x)
x = Conv2D(128, (3, 3), strides=(1, 1), padding="same", activation="relu", name="7th_conv")(x)
x = Conv2D(128, (3, 3), strides=(1, 1), padding="same", activation="relu", name="8th_conv")(x)
x = MaxPooling2D((2, 2), strides=(2, 2), padding="valid", name="3rd_pool")(x)
f = int(np.floor(crop_size/8))
sk_out = Conv2D(crop_size*crop_size, (f, f), strides=(1, 1), padding="valid", activation="sigmoid", name="flatten")(x)
sk_model = Model(inputs=sk_in, outputs=sk_out)
sk_model.summary()
```

```python
# Fit model
sk_model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
sk_model.fit(X_train, Y_train, batch_size=32, epochs=3)
```

```python
# Evaluate model
loss, accuracy = sk_model.evaluate(X_test, Y_test, batch_size=32)
print("Accuracy on test set is ", accuracy)
```

```python
#Predict for example of test set
example = 1
coord_pred = sk_model.predict(X_train[example, :, :, :].reshape(1, crop_size, crop_size, 1))
# Plotting original neurons
plt.imshow(sk_plan[example, :, :])
plt.colorbar()
plt.title("Final plan with artificially created neurons (detail)")
# Adding centers of neurons
sk_coord2 = sk_coord[sk_coord[:, 0] == example, 1:3]
plt.scatter(sk_coord2[:, 1], sk_coord2[:, 0], s=5, c=[1, 0, 0], marker='.')
# Adding prediction
coord_pred = coord_pred.reshape(crop_size, crop_size)
#plt.imshow(coord_pred)
#plt.colorbar()
pred_x, pred_y = np.where(coord_pred == 1)
plt.scatter(pred_x, pred_y, s=5, c=[0, 0, 0], marker='.')
print(coord_pred)
```

<br><br><br><br>We do have a big problem: since 97% of the output is class 0, the algorithm with adam classifies all the outputs as 0, resulting in an accuracy of 97%, but not detecting centers of neurons, as we wanted. After searching for outlier/anomaly detection from the internet, I found that the use of autoencoders may be a solution. Here are two papers about it:
https://togaware.com/papers/dawak02.pdf
https://cds.cern.ch/record/2209085/files/Outlier%20detection%20using%20autoencoders.%20Olga%20Lyudchick%20(NMS).pdf
What I am going to try is switch to a multiclass classification with random classes for all the pixels which are not labeled centers of neurons. This was, as it will be impossible for the algorithm to accurately classify them, it might focus on the centers of the neurons instead and give us good results.

We are going to use 40 classes, as centers of neurons represent 2.5% of total outputs.

```python
Y_train2 = np.copy(Y_train.reshape(Y_train.shape[0], Y_train.shape[3]))
ytrainx, ytrainy = np.where(Y_train2 == 0)
s_ytrain = ytrainx.shape[0]
Y_train2[ytrainx, ytrainy] = np.random.random_integers(2, 40, s_ytrain)
Y_train2 = Y_train2.reshape(Y_train2.shape[0], 1, 1, Y_train2.shape[1])
Y_test2 = np.copy(Y_test.reshape(Y_test.shape[0], Y_test.shape[3]))
ytestx, ytesty = np.where(Y_test2 == 0)
s_ytest = ytestx.shape[0]
Y_test2[ytestx, ytesty] = np.random.random_integers(2, 40, s_ytest)
Y_test2 = Y_test2.reshape(Y_test2.shape[0], 1, 1, Y_test2.shape[1])
```

```python
# Building neural network
sk_in = Input(shape=(crop_size, crop_size, 1))
x = Conv2D(32, (3, 3), strides=(1, 1), padding="same", activation="relu", name="1st_conv")(sk_in)
x = Conv2D(32, (3, 3), strides=(1, 1), padding="same", activation="relu", name="2nd_conv")(x)
x = MaxPooling2D((2, 2), strides=(2, 2), padding="valid", name="1st_pool")(x)
x = Conv2D(64, (3, 3), strides=(1, 1), padding="same", activation="relu", name="3rd_conv")(x)
x = Conv2D(64, (3, 3), strides=(1, 1), padding="same", activation="relu", name="4th_conv")(x)
x = Conv2D(64, (3, 3), strides=(1, 1), padding="same", activation="relu", name="5th_conv")(x)
x = MaxPooling2D((2, 2), strides=(2, 2), padding="valid", name="2nd_pool")(x)
x = Conv2D(128, (3, 3), strides=(1, 1), padding="same", activation="relu", name="6th_conv")(x)
x = Conv2D(128, (3, 3), strides=(1, 1), padding="same", activation="relu", name="7th_conv")(x)
x = Conv2D(128, (3, 3), strides=(1, 1), padding="same", activation="relu", name="8th_conv")(x)
x = MaxPooling2D((2, 2), strides=(2, 2), padding="valid", name="3rd_pool")(x)
sk_out = Conv2D(crop_size*crop_size, (f, f), strides=(1, 1), padding="valid", activation="sigmoid", name="flatten")(x)
sk_model = Model(inputs=sk_in, outputs=sk_out)
```
